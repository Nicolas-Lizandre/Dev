```python
#Used last Lab (Lab1)

def clean_line(line):
  cleaned_chars = []
  ponctuation = "*!=_:/;.,?()[]{'\"-"
  numbers = "0123456789"
  for letter in line:
    if 'A' <= letter <= 'Z':
      cleaned_chars.append(letter.lower())
    elif letter in ponctuation:
      cleaned_chars.append(" ") #replace ponctuation by a space
    elif letter in numbers:
      cleaned_chars.append(" ")
    else:
      cleaned_chars.append(letter)
  #removing space

  return "".join(cleaned_chars)


unique_words = log_file.flatMap(lambda x : clean_line(x.page).split(" ")).filter(lambda line : line.strip()!= "").distinct().cache()
unique_words.take(10)
```

```python
from pyspark.sql.functions import udf, split, explode, trim
from pyspark.sql.types import StringType

clean_udf = udf(clean_line, StringType())
```

```python
df_unique_words = (
    df_file
    .withColumn("clean_title", clean_udf(df_file.page))
    .select(explode(split(trim("clean_title"), " ")).alias("word"))
    .filter("word != ''")
    .distinct()
)

df_unique_words.show(5)
```

```python
# Register the Python clean_line function as a Spark SQL UDF
spark.udf.register("clean_line", clean_line, StringType())

unique_words = spark.sql("SELECT DISTINCT word FROM (SELECT explode(split(clean_line(page), ' ')) AS word FROM Table) WHERE word <> '' ")
unique_words.show(5)
```
