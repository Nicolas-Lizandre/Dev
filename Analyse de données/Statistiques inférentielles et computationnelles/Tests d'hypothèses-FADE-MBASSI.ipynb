### Question 1 : Donner une estimation de la moyenne "muEstim", de la variance "varEstim" de la loi suivie par les $X_i$.

```python
def moy_emp(data):
    s=0
    for i in range(len(data)):
        s=s+data[i]
    return s/len(data)

def var_emp_2B(data):
    s=0
    for i in range(len(data)):
        s=s+data[i]**2
    return s/(len(data)) - moy_emp(data)**2

muEstim = moy_emp(data)
varEstim = var_emp_2B(data)

print("MoyenneEstim : "+str(np.mean(data)))
print("MoyenneEstim (2B) : "+str(moy_emp(data)))
print("VarianceEstim : "+str(np.var(data, ddof=1)))
print("VarianceEstim (2B): "+str(var_emp_2B(data)))
```

### Question 6 : Créer une fonction "fR(x)" qui renvoie la valeur de la fonction de répartition d'une loi normale de moyenne "muEstim" et de variance "varEstim" en $x$. On pourra utiliser le calcul d'intégrale par la fonction "quad" de "scipy.integrate".

```python
def fR(x):
    densite = lambda t: 1/(np.sqrt(2*np.pi*varEstim)) * np.exp(-(t-muEstim)**2/(2*varEstim))
    result, _ = quad(densite, -np.inf, x)
    return result

print(f"fR(10) = {fR(10):.4f}")
print(f"fR(muEstim) = {fR(muEstim):.4f}")
```

### Question 9 : Créer une fonction "regroupeGauche(liste,k)" qui prend en entrée une liste et un entier positif $k$ et retourne la liste obtenue en regroupant et en sommant les $k$ premiers éléments de la liste.

```python
def regroupeGauche(liste, k):
    return [sum(liste[:k])] + liste[k:]
```

### Question 10 : Faire de même à droite en créant la fonction "regroupeDroite(liste,k)".

```python
def regroupeDroite(liste, k):
    return liste[:-k] + [sum(liste[-k:])]
```

### Question 14 : Définir une fonction "skewnorm(x,e,w,a)" qui donne la densité en $x$ d'une loi normale asymétrique de moyenne $e$, de variance $w$ et de coefficient d'asymétrie $a$ (compiler le code ci-dessous). Représenter cette densité avec par exemple $e=0, w=1$ et $a=4$.

```python
def pdf(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-x**2/2)
def cdf(x):
    return (1 + erf(x/np.sqrt(2))) / 2

def skewnorm(x,m,s,a):
    t = (x-m) / s
    return 2 / s * pdf(t) * cdf(a*t)
```

```python
# Superposition des densités : loi normale centrée réduite N(0,1) et skewnorm(0,1,4)
x = np.linspace(-4, 6, 500)

plt.figure(figsize=(10, 6))
plt.plot(x, skewnorm(x, 0, 1, 0), label='Loi normale centrée réduite (a=0)')
plt.plot(x, skewnorm(x, 0, 1, 4), label='Loi normale asymétrique (m=0, s=1, a=4)')
plt.xlabel('x')
plt.ylabel('Densité')
plt.legend()
plt.title('Comparaison loi normale et loi normale asymétrique')
plt.grid(True)
plt.show()
```

### Question 19 : Créer la matrice de contingence des fréquences observées et celle des fréquences théoriques sous l'hypothèse d'indépendance. Les classes seront les suivantes. \begin{align*}
[0,10.5[;[10.5,13[;[13,15[;[15,16[,[16,20].\end{align*}
### Calculer la distance du $\chi^2$ de ce test d'indépendance, puis conclure (test unilatéral gauche) en précisant la valeur critique et le nombre de degrés de liberté au seuil de risque de première espèce de $5\%$.

```python
bornes = [0, 10.5, 13, 15, 16, 20]

# Fonction pour compter les effectifs par classe
def compter_classes(data, bornes):
    effectifs = []
    for i in range(len(bornes) - 1):
        count = np.sum((np.array(data) >= bornes[i]) & (np.array(data) < bornes[i+1]))
        effectifs.append(count)
    return effectifs

# Effectifs observés par groupe
effFemmes = compter_classes(dataFemmes, bornes)
effHommes = compter_classes(dataHommes, bornes)

print("Effectifs Femmes par classe:", effFemmes)
print("Effectifs Hommes par classe:", effHommes)

# Matrice de contingence observée
matContingenceObs = np.array([effFemmes, effHommes])
print("\nMatrice de contingence observée:")
print(matContingenceObs)
```

```python
nFemmes = len(dataFemmes)
nHommes = len(dataHommes)
nTotal = nFemmes + nHommes

print(f"Nombre de femmes: {nFemmes}")
print(f"Nombre d'hommes: {nHommes}")
print(f"Total: {nTotal}")

# Effectifs marginaux par classe (somme par colonne)
margClasses = np.array(effFemmes) + np.array(effHommes)
print(f"\nEffectifs marginaux par classe: {margClasses}")

# Matrice des effectifs théoriques sous H0 (indépendance)
# E_ij = (n_i. * n_.j) / n
effThFemmes = [nFemmes * marg / nTotal for marg in margClasses]
effThHommes = [nHommes * marg / nTotal for marg in margClasses]

matContingenceTh = np.array([effThFemmes, effThHommes])
print("\nMatrice de contingence théorique (sous H0: indépendance):")
print(np.round(matContingenceTh, 2))
```

```python
# Calcul de la distance du chi2
chi2_indep = np.sum((matContingenceObs - matContingenceTh)**2 / matContingenceTh)
print(f"Distance chi2 = {chi2_indep:.4f}")

# Degrés de liberté : (nb_lignes - 1) * (nb_colonnes - 1)
nb_lignes = 2
nb_colonnes = len(bornes) - 1
ddl_indep = (nb_lignes - 1) * (nb_colonnes - 1)
print(f"Degrés de liberté : ({nb_lignes}-1) x ({nb_colonnes}-1) = {ddl_indep}")

# Valeur critique à 5%
from scipy.stats import chi2 as chi2_dist
valeur_critique_indep = chi2_dist.ppf(0.95, ddl_indep)
print(f"Valeur critique (alpha=5%) : {valeur_critique_indep:.4f}")
```
